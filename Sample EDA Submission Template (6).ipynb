````python
# To get overall statitics of a perticular function
class  AreaCode:

    try:

        def __init__(self, data):
            self.data = data

        def describe(self,AreaCode):
            return self.data[self.data['Area code']==AreaCode].describe()

        def mean(self,AreaCode):
            Mean_df=pd.DataFrame(self.data[self.data['Area code']==AreaCode].describe().loc['mean'])
            return Mean_df.rename(columns={AreaCode:'mean'})

        def median(self,AreaCode):
            Mean_df=pd.DataFrame(self.data.groupby('Area code').median().loc[AreaCode])
            return Mean_df.rename(columns={AreaCode:'median'})



    except:
        print("Wrong area code")

````
