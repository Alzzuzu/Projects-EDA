````python
#Write your code to make your dataset analysis-ready.
# Create a copy of the current dataset and assign a short variable name.
df=telecom_df.copy()
#Check the number of churning customers from the given data.
#It helps to understand the depth of the issue.
print("No. of customers Churning : -",len(df[df['Churn']==True]))
#Assigning churn customers' and other's data to a variable for easy access.
df_churn=df[(df['Churn']==True)]
df_nonchurn=df[(df['Churn']==False)]
````

````python
# we collect number of churn statewise
churn_by_state = df.groupby('State')['Churn'].value_counts().unstack()
Sorted_churn_by_state=churn_by_state.sort_values(by=True , ascending=False ,)
print(Sorted_churn_by_state)
print(f'Top 5 states with most number of churn are {Sorted_churn_by_state.index[0:5].tolist()} ')
````

````python
# we have only 3 area codes are present in data
#churn data by area number
churn_by_area = df.groupby('Area code')['Churn'].value_counts().unstack().sort_values(by=True,ascending=False)
sorted_churn_by_area_reset = churn_by_area.reset_index()

# churn percentage per area
sorted_churn_by_area_reset['churn percentage ']=100*sorted_churn_by_area_reset[True]/(sorted_churn_by_area_reset[True]+sorted_churn_by_area_reset[False])
print(sorted_churn_by_area_reset)
````

````python
# To get overall statitics of a perticular function
class  AreaCode:

    try:

        def __init__(self, data):
            self.data = data

        def describe(self,AreaCode):
            return self.data[self.data['Area code']==AreaCode].describe()

        def mean(self,AreaCode):
            Mean_df=pd.DataFrame(self.data[self.data['Area code']==AreaCode].describe().loc['mean'])
            return Mean_df.rename(columns={AreaCode:'mean'})

        def median(self,AreaCode):
            Mean_df=pd.DataFrame(self.data.iloc[:,:19].groupby('Area code').median().loc[AreaCode])
            return Mean_df.rename(columns={AreaCode:'median'})



    except:
        print("Wrong area code")

````

````python
# Getting statistics for each area code.
# For areacode 408
Discription_408=AreaCode(df).describe(408)
Mean_408=AreaCode(df).mean(408)
Median_408=AreaCode(df).median(408)
````

````python
#Forarea code 510
Discription_510=AreaCode(df).describe(510)
Mean_510=AreaCode(df).mean(510)
Median_510=AreaCode(df).median(510)
````

````python
# For area code 415
Discription_415=AreaCode(df).describe(415)
Mean_415=AreaCode(df).mean(415)
Median_415=AreaCode(df).median(415)
````

````python
#getting relation between defferent price rate of churn customers
#usage and rate
usage_rate=df_churn[['Total day minutes','Total day charge','Total eve minutes','Total eve charge','Total night minutes','Total night charge']]
# adding rate per mint (R/M) for each time
# For day time
usage_rate['day R/M']=usage_rate['Total day charge']/usage_rate['Total day minutes']
# For evening
usage_rate['eve R/M']=usage_rate['Total eve charge']/usage_rate['Total eve minutes']
# For night
usage_rate['night R/M']=usage_rate['Total night charge']/usage_rate['Total night minutes']
````

````python
#Finding mean for usage and rate for each time of churn
Mean_usage_rate_non=pd.DataFrame(usage_rate_non.mean()).rename(columns={0:'mean nonchurn'})
Mean_usage_rate_non
````
